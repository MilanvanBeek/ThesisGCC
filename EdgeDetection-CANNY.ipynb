# CANNY TESTS

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
Image = cv2.imread('output-onlinepngtools.png', cv2.IMREAD_GRAYSCALE)
plt.subplot(121),plt.imshow(Image,cmap = 'gray')
```

```python
def auto_canny(image, sigma=0.99):
    # compute the median of the single channel pixel intensities
    v = np.median(image)
    # apply automatic Canny edge detection using the computed median
    lower = 20 #/ 20
    upper = 35 #/ 60
    edged = cv2.Canny(image, lower, upper)
    # return the edged image
    return edged
```

```python
edged = auto_canny(Image)
```

```python
edged
```

```python
plt.subplot(121),plt.imshow(Image,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edged,cmap = 'gray')
plt.title('Image UT:20 LT:35'), plt.xticks([]), plt.yticks([])
plt.savefig('Image UT,20 LT,35.jpg')
plt.show()
```

# Threshold SLIDER

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt 

def callback(x):
    print(x)

img = cv2.imread('output-onlinepngtools.png', 0) #read image as grayscale


canny = cv2.Canny(img, 85, 255) 

cv2.namedWindow('image') # make a window with name 'image'
cv2.createTrackbar('L', 'image', 0, 255, callback) #lower threshold trackbar for window 'image
cv2.createTrackbar('U', 'image', 0, 255, callback) #upper threshold trackbar for window 'image

while(1):
    numpy_horizontal_concat = np.concatenate((img, canny), axis=1) # to display image side by side
    cv2.imshow('image', numpy_horizontal_concat)
    k = cv2.waitKey(1) & 0xFF
    if k == 27: #escape key
        break
    l = cv2.getTrackbarPos('L', 'image')
    u = cv2.getTrackbarPos('U', 'image')

    canny = cv2.Canny(img, l, u)

cv2.destroyAllWindows()
```
